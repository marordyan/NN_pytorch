```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import rcParams

rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [12, 12]
```

```python
## Simulate the Lorenz System

dt = 0.01
T = 80
t = np.arange(0,T+dt,dt)
beta = 8/3
sigma = 10
rho = 28
n_samples = 500

nn_input = np.zeros((n_samples*(len(t)-1),3))
nn_output = np.zeros_like(nn_input)

fig,ax = plt.subplots(1,1,subplot_kw={'projection': '3d'})


def lorenz_deriv(x_y_z, t0, sigma=sigma, beta=beta, rho=rho):
    x, y, z = x_y_z
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

np.random.seed(123)
x0 = -15 + 30 * np.random.random((n_samples, 3))

x_t = np.asarray([integrate.odeint(lorenz_deriv, x0_j, t)
                  for x0_j in x0])



for j in range(n_samples):
    # nn_input[j*(len(t)-1):(j+1)*(len(t)-1),:] = x_t[j,:-1,:]
    # nn_output[j*(len(t)-1):(j+1)*(len(t)-1),:] = x_t[j,1:,:]
    x, y, z = x_t[j,:,:].T
    x = (x-np.mean(x))/np.std(x)
    y = (y-np.mean(y))/np.std(y)
    z = (z - np.mean(z))/np.std(z)
    nn_input[j*(len(t)-1):(j+1)*(len(t)-1),:] = (np.array([x[:-1],y[:-1],z[:-1]])).T
    nn_output[j*(len(t)-1):(j+1)*(len(t)-1),:] = (np.array([x[1:],y[1:],z[1:]])).T
    ax.plot(x, y, z,linewidth=1)
    #ax.scatter(x0[j,0],x0[j,1],x0[j,2],color='r')
    ax.scatter(x[0],y[0],z[0],color='r')

# nn_input = (nn_input - np.mean(nn_input))/(np.std(nn_input))
# nn_output = (nn_output - np.mean(nn_output))/(np.std(nn_output))
             
ax.view_init(18, -113)
plt.show()
```

```python
## NN

import torch
from torch import nn
from torch import optim

# def model(nn.Module):
#     __init__(gpu)

model = nn.Sequential(nn.Linear(3, 10),
                    nn.ReLU(),
                    nn.Linear(10, 10),
                    nn.ReLU(),
                    nn.Linear(10, 10),
                    nn.ReLU(),
                    nn.Linear(10,3),
                    )

#model.to(device=torch.device('cuda:4'))

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr = 0.001)
```

```python
my_device = 'cpu'
epochs = 5000
training_loss = np.zeros(epochs)
batch_size = 50
x_norm = (x_t - x_t.mean((0,1)))/x_t.std((0,1))
for e in range(epochs):
    running_loss = 0
    for i in range(int(n_samples/batch_size)):
        idx = np.random.choice(n_samples, batch_size, replace = False)
        curr_input = torch.from_numpy((x_norm[idx,:-1,:]).reshape(-1,3)).to(device = my_device)
        curr_output = torch.from_numpy((x_norm[idx,1:,:]).reshape(-1,3)).to(device = my_device)
        optimizer.zero_grad()
        output = model(curr_input.float())
        loss = criterion(output, curr_output.float())
        loss.backward()
        optimizer.step()
        running_loss += loss
    training_loss[e] = running_loss/int(n_samples/batch_size)
    if not(e%100):
        print(f"Training loss: {training_loss[e]}")    

torch.save(model,'./lorenz_model.pth')
```

```python
nn_input = torch.from_numpy(nn_input.astype(np.float64))
nn_output = torch.from_numpy(nn_output.astype(np.float64))
```

```python
#### Test model on training set

model.eval()
with torch.no_grad():
    output = model.forward(nn_input.float())
    loss = criterion(output,nn_output)
    print(f"Test loss on training data: {loss}")
    print(f"Manually calculated loss: {np.sum(((output - nn_output).numpy())**2)/((nn_input.shape[0]*nn_input.shape[1]))}")
    plt.plot(output[0:10000000, 1], label = 'eval_output')
    plt.plot(nn_output[0:10000000, 1], label = 'output')
    #plt.plot(y_train[0:1000,1],label='predict')
    plt.show()
```

```python
## Generate test data

fig,ax = plt.subplots(1,1,subplot_kw={'projection': '3d'})
np.random.seed(12)
x0 = -15 + 30 * np.random.random((n_samples, 3))

x_test = np.asarray([integrate.odeint(lorenz_deriv, x0_j, t)
                  for x0_j in x0])

n_test_samples = 5

for j in range(n_test_samples):
    x, y, z = x_test[j,:,:].T
    x = (x-np.mean(x))/np.std(x)
    y = (y-np.mean(y))/np.std(y)
    z = (z - np.mean(z))/np.std(z)
    ax.plot(x, y, z,linewidth=1)
    ax.scatter(x[0],y[0],z[0],color='r')
             
ax.view_init(18, -113)
plt.show()
```

```python
y_train = torch.zeros(x_test.shape[1],x_test.shape[2])
test_input = torch.from_numpy((x_test[0,:-1,:] - x_test.mean((0,1)))/x_test.std((0,1)))
test_output = torch.from_numpy((x_test[0,1:,:] - x_test.mean((0,1)))/x_test.std((0,1)))
y_train[0,:] = test_input[0,:]


#print(f'y_train[0] is now: {y_train[0,:]}')
with torch.no_grad():
    running_loss = 0
    man_loss = 0
    for j in range(int(T/dt)):
        #print(f'y_train[{j}] is now: {y_train[j,:]}')
        if j < 20:
            y_train[j+1,:] = model(test_input[j,:].float())
        else:
            y_train[j+1,:] = model(y_train[j,:])
        loss = criterion(y_train[j+1,:], test_output[j,:])
        #print(f'expected output for y_train[{j+1}] is {nn_output[j,:]}')

        running_loss += loss.item()
        if not(j%100):
            print(f'Running loss is: {running_loss/((j+1))}')
            #print(f'Manual loss is: {man_loss/((j+1)*3)}')
```

```python
plt.plot(test_input[0:1000, 1], label = 'input')
plt.plot(test_output[0:1000, 1], label = 'output')
plt.plot(y_train[0:1000,1],label='predict')
plt.show()
```
